```python
import keras
#from keras.datasets import cifar10
from keras.datasets import mnist 
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential, load_model
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras import optimizers
from keras.callbacks import EarlyStopping, ModelCheckpoint
import numpy 
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

%matplotlib inline

def plot_data(history, title=''):
    plt.clf()
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
    fig.suptitle(title)

    # Plot training & validation accuracy values
    ax1.plot(history.history['acc'])
    ax1.plot(history.history['val_acc'])
    #ax1.set_title('Model accuracy')
    ax1.set_ylabel('Accuracy')
    ax1.set_xlabel('Epoch')
    ax1.legend(['Train', 'Test'], loc='upper left')
    ax1.xaxis.set_major_locator(MaxNLocator(integer=True))
    #ax1.set_ylim(0.0, 1.0)
    
    # Plot training & validation loss values
    ax2.plot(history.history['loss'])
    ax2.plot(history.history['val_loss'])
    #ax2.set_title('Model loss')
    ax2.set_ylabel('Loss')
    ax2.set_xlabel('Epoch')
    ax2.legend(['Train', 'Test'], loc='upper left')
    ax2.xaxis.set_major_locator(MaxNLocator(integer=True))
    
    plt.show()
```

```python
def build_default_model():
    # Step：選擇模型, 順序模型是多個網絡層的線性堆疊
    model = Sequential()

    # Step：構建網絡層
    model.add(Dense( 500,input_shape=(784,))) # 輸入層，28*28=784   
    model.add(Activation('relu')) # 激活函數是relu   

    model.add(Dense( 500)) # 隱藏層節點500個   
    model.add(Activation('relu'))  

    model.add(Dense( 500)) # 隱藏層節點500個   
    model.add(Activation('relu'))  

    model.add(Dense( 500)) # 隱藏層節點500個   
    model.add(Activation('relu'))  
   
    model.add(Dense( 10)) # 輸出結果是10個類別，所以維度是10   
    model.add(Activation('softmax')) # 最後一層用softmax作為激活函數

    return model
```

```python
def build_simple_model():
    # Step：選擇模型, 順序模型是多個網絡層的線性堆疊
    model = Sequential()

    # Step：構建網絡層
    model.add(Dense( 32,input_shape=(784,))) # 輸入層，28*28=784   
    model.add(Activation('relu')) # 激活函數是relu   

    model.add(Dense( 32)) # 隱藏層節點32個   
    model.add(Activation('relu'))  

    model.add(Dense( 10)) # 輸出結果是10個類別，所以維度是10   
    model.add(Activation('softmax')) # 最後一層用softmax作為激活函數
    
    # 模型建立完成後，統計參數總量
    #print("Total Parameters：%d" % model.count_params())
    
    # 輸出模型摘要資訊
    #print(model.summary())
    
    return model
```

```python
'''
ref: Day_046
'''
# plot error images
def plot_error_images(x_test, y_test_int, y_pred_int, max_display=20):
    y_test = y_test_int
    y_pred = y_pred_int
    
    #y_pred_err_inx = [i for i in range(len(y_pred)) if not np.array_equal(y_pred[i],  y_test[i])]
    y_pred_err_inx = [i for i in range(len(y_pred)) if y_pred[i] != y_test[i]]
    x_test_err = x_test[y_pred_err_inx]
    y_test_err = y_test[y_pred_err_inx]
    y_pred_err = y_pred[y_pred_err_inx]

    plt.clf()
    # Plot the prediction
    fig = plt.figure(figsize=(30, 30))  # figure size in inches
    fig.suptitle(f"Predict error data: count={len(x_test_err)}\nblack:true value\nred:predict value", horizontalalignment='center', verticalalignment='top', fontsize=28) 
    #fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.3, wspace=0.1)
    fig.subplots_adjust(hspace=0.3, wspace=0.1)

    display_cnt = np.minimum(max_display, len(y_pred_err))
    cols = 5
    rows = (display_cnt // cols) + 1

    # plot the digits: each image is 28x28 pixels
    for i in range(display_cnt):
        ax = fig.add_subplot(rows, cols, i + 1, xticks=[], yticks=[])
        ax.set_title(f'#{i+1}', fontsize=28)

        ax.imshow(x_test_err.reshape(-1, 28, 28)[i], cmap=plt.cm.binary,
                  interpolation='nearest')

        # label the image with the target value
        #ax.text(0, 1, str(y_test_err[i]), color='black')
        #ax.text(0, 7, str(y_pred_err[i]), color='red')
        ax.text(0,  2, str(y_test_err[i]), fontsize=20, color='black')
        ax.text(0, 27, str(y_pred_err[i]), fontsize=20, color='red')
                   
    plt.show()
```

```python
import numpy as np

# Step：輸出
def output_accuracy(model, X_test, Y_test):
    print(" test set " )
    scores = model.evaluate(X_test,Y_test,batch_size=200,verbose= 0)
    print("")
    print(f"The test (loss, accuracy) is {scores}")
    result = model.predict(X_test,batch_size=200,verbose= 0)

    #print(result[:5])
    #print(Y_test[:5])

    result_max = numpy.argmax(result, axis = 1)
    test_max = numpy.argmax(Y_test, axis = 1)

    result_bool = numpy.equal(result_max, test_max)
    true_num = numpy.sum(result_bool)
    print(f"The accuracy of the model is {(true_num/len(result_bool))}")
    
    y_pred = result_max
    y_test = test_max
    plot_error_images(X_test, y_test, y_pred)
    
```

```python
# Step：資料分割
# 使用Keras自帶的mnist工具讀取數據（第一次需要聯網）
(X_train, y_train), (X_test, y_test) = mnist.load_data() 
print(f'X_train={X_train.shape}')
print(f'y_train={y_train.shape}')
print(f'X_test={X_test.shape}')
print(f'y_test={y_test.shape}')

# 由於mist的輸入數據維度是(num, 28 , 28)，這裡需要把後面的維度直接拼起來變成784維   
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1] * X_train.shape[2])
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1] * X_test.shape[2])  
Y_train = (numpy.arange(10) == y_train[:, None]).astype(int)
Y_test = (numpy.arange(10) == y_test[:, None]).astype(int)

print(f'X_train={X_train.shape}')
print(f'y_train={y_train.shape}')
print(f'X_test={X_test.shape}')
print(f'y_test={y_test.shape}')

print(f'y_train[:5]={y_train[:5]}')
print(f'Y_train[:5]=\n{Y_train[:5]}')
print(f'y_test[:5]={y_test[:5]}')
print(f'Y_test[:5]=\n{Y_test[:5]}')
```

### 調整 batch_size, epoch , 觀察accurancy, loss 的變化

```python
'''
   Step: 編譯
   Step: 訓練
   .fit的一些參數
   batch_size：對總的樣本數進行分組，每組包含的樣本數量
   epochs ：訓練次數
   shuffle：是否把數據隨機打亂之後再進行訓練
   validation_split：拿出百分之多少用來做交叉驗證
   verbose：屏顯模式 - 0：不輸出, 1：輸出進度, 2：輸出每次的訓練結果
''' 
optimizer = 'adam'
#batch_size_list = [50, 128, 256] 
#epochs_list = [10, 30]
batch_size_list = [128] 
epochs_list = [20]
for batch_size in batch_size_list:
    for epochs in epochs_list:
        title = f"optimizer='{optimizer}'  batch_size={batch_size}  epochs={epochs}"
        print(title)

        model = build_default_model()
        print(model.summary())

        # Step: 編譯
        # 多分類問題, 改用 categorical_crossentropy
        model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
        #model.compile(optimizer = optimizer, loss = 'categorical_crossentropy', metrics = ['accuracy'])

        # Step: 訓練
        history = model.fit(X_train,Y_train, batch_size=batch_size,epochs=epochs,shuffle=True,verbose=2,validation_split=0.3)
        
        plot_data(history, title=title)
```

```python
app_train.describe()
```

```python
df_numbers = app_train.select_dtypes(include=['number'])
for col_name in df_numbers.columns.values:
    print(f"欄位[{col_name}] : 平均數=[{app_train[col_name].mean():.2f}]\t標準差=[{app_train[col_name].std():.2f}]")
```

```python
df_norm = (app_train['AMT_GOODS_PRICE']-app_train['AMT_GOODS_PRICE'].mean())/(app_train['AMT_GOODS_PRICE'].max()-app_train['AMT_GOODS_PRICE'].min())
df_norm.hist(bins=100)
```

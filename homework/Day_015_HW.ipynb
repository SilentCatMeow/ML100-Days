3. 請用 numpy 建立一個 1000 x 3, 數值分布為常態分佈的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
#plt.style.use('ggplot')
plt.style.use('seaborn')

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
nrow = 10
ncol = 10
matrix = (np.random.random(nrow*ncol)*2-1).reshape(nrow,ncol)

plt.figure(figsize=(10,10))

heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, annot = True)

plt.show()
```

請用 numpy 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

```python
#sns.set(style="ticks", color_codes=True)

nrow = 1000
ncol = 3
matrix = (np.random.random(nrow*ncol)*2-1).reshape(nrow,ncol)

indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)


grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)

grid.map_upper(sns.scatterplot , alpha = 0.2)
grid.map_diag(plt.hist, bins=50)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```

請用 numpy 建立一個 1000 x 3, 數值分布為常態分佈的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

```python
nrow = 1000
ncol = 3
matrix = np.random.randn(nrow*ncol).reshape(nrow,ncol)

indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)


grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)

grid.map_upper(sns.scatterplot , alpha = 0.2)
grid.map_diag(plt.hist, bins=50)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```

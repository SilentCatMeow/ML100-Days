## 線性整流函數（Rectified Linear Unit, ReLU）
$ ReLU(x) = \left\{
  \begin{array}{lr}
    x & : x \gt 0\\
    0 & : x \le 0
  \end{array}
\right. $

$ ReLU(x) = max(0, x) $

## 線性整流函數一階導數（derivative ReLU）
$ dReLU(x) = \left\{
  \begin{array}{lr}
    1 & : x \gt 0\\
    0 & : x \le 0
  \end{array}
\right. $

```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

#from matplotlib import rc
#rc('text', usetex=True)
#rc['text.usetex'] = True

#from matplotlib import rcParams
#rcParams['text.usetex'] = True

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''

import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def ReLU(x):
    """ReLU returns x if x>0, else 0."""
    return np.maximum(0, x)

# ref : Day 072 pdf
#def ReLU(x):
#    """ReLU returns x if x>0, else 0."""
#    return abs(x) * (x > 0)

#------------------------------------------------------
## not avalibale for array input
#def dReLU(x):
#    """dReLU returns 1 if x>0, else 0."""
#    return (1 if x > 0 else 0)

#def dReLU(x):
#    """dReLU returns 1 if x>0, else 0."""
#    return map(lambda x: 1 if x > 0 else 0, x)

# ref : Day 072 pdf
def dReLU(x):
    """dReLU returns 1 if x>0, else 0."""
    return (1 * (x > 0))

# ref: https://stackoverflow.com/questions/45021963/relu-prime-with-numpy-array
#def dReLU(x):
#    return np.where(x > 0, 1.0, 0.0)

#def dReLU(x):
#    return (x>0).astype(x.dtype)

#def dReLU(x):
#    return np.heaviside(x, 0)
```

```python
#x = plt.linspace(-10, 10, 100)
x = np.arange(-10, 10, 0.01)

plt.plot(x, ReLU(x), 'r')

#plt.plot(x, dReLU(x), 'b')
plt.step(x, dReLU(x), 'b')

plt.show()
```

```python
X  = [x for x in range(-10, 10)]
Y  = [ReLU(x) for x in X]
dY = [dReLU(x) for x in X]


plt.plot(X, Y, 'r')
plt.title('ReLU Function')
plt.legend([r'$ ReLU(x) = max(0, x) $'], loc='best', fontsize=12)
plt.grid()
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))
plt.show()

# error
#plt.plot(X, dY, 'b')
plt.step(X, dY, 'b')
plt.title('derivative ReLU Function')
plt.legend(['1: x>0 \n0: x<=0'], loc='best', fontsize=12)
plt.grid(color='b', linewidth='0.3', linestyle='--')
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))

plt.show()
```

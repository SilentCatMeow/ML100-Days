```python
import os
import numpy as np
import pandas as pd

#print("Pandas version:"+pd.__version__)
# Pandas version:0.23.4
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
#print(f'Columns: {app_train.columns.size}')
print(f'Rows: {app_train.shape[0]}')
print(f'Columns: {app_train.shape[1]}')
```

#### 列出所有欄位

```python
col_names = ', '.join(app_train.columns)
print('Column names: [\n%s\n]' % (col_names))
```

#### 截取部分資料

```python
app_train.head(n = 2)
```

```python
#print(f'tail 2: \n{app_train.tail(n = 2)}\n')
app_train.tail(n = 2)
```

```python
app_train.sample(n = 10)
```
